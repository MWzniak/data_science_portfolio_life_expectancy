```python
# Importing the libraries for exploratory data analysis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
dataset=pd.read_csv('life_expectancy.csv')
```

```python
# Examining the preview of the dataset
dataset.sample(n=5)
```

```python
# Gathering information about the null values and dtypes of variables
dataset.info()
# Some null values are present
```

```python
# Examining basic descriptive statistics data
dataset.describe().T
```

```python
# Counting the missing data
def null_counter(columns,dataset):
    col_names=list(columns)
    counts=[]
    for n in col_names:
        counts.append(dataset[n].isnull().sum())
    null_counts=pd.DataFrame({'col_name':col_names,'null_count':counts})
    null_counts=null_counts[null_counts.null_count!=0]
    return null_counts
null_counts=null_counter(dataset.columns,dataset)
null_counts
```

```python
# Dropping rows with missing values which are just a tiny fraction of the whole column
for n in dataset.columns:
    if dataset[n].isnull().sum()<40:
        dataset=dataset.dropna(subset=n)
null_counter(dataset.columns,dataset)
```

```python
# This took care of some of the other missing values as well, however there are too many to simply drop them. These have to be imputed.
# Examining correlation between different variables
plt.figure(figsize=(15,15))
sns.heatmap(dataset.corr(numeric_only=True),cmap='vlag',annot=True)
plt.show()
```

```python
# 'Alcohol' feature is correlated with some other features, but these missing values are less than 10% of the whole dataset. These can be replaced with a mean value.
dataset['Alcohol']=dataset['Alcohol'].fillna(dataset['Alcohol'].mean())
null_counter(dataset.columns,dataset)
```

```python
# 'Hepatitis B' missing values should be predicted, because it is highly correlated with some features and they make up about 20% of the whole dataset
from sklearn.linear_model import LinearRegression
dataset=dataset.reset_index(drop=True)
test_dataset=dataset[dataset['Hepatitis B'].isnull()]
train_dataset=dataset[dataset['Hepatitis B'].notnull()]
y_train=train_dataset['Hepatitis B']
X_train=train_dataset.drop(['Hepatitis B','Country','Status','Total expenditure','GDP', 'Population','Income composition of resources', 'Schooling'],axis=1)
X_test=test_dataset.drop(['Hepatitis B','Country','Status','Total expenditure','GDP', 'Population','Income composition of resources', 'Schooling'],axis=1)
linreg=LinearRegression()
linreg.fit(X_train,y_train)
y_pred=linreg.predict(X_test)
```

```python
# Replacing missing values
i_list=[]
for n in enumerate(dataset['Hepatitis B'].isnull()):
    if n[1]:
        i_list.append(n[0])
        m=0
for n in i_list:
    dataset['Hepatitis B'][n]=y_pred[m]
    m+=1
null_counter(dataset.columns,dataset)
```

```python
# 'Total expenditure' feature is not really correlated with other features. These missing values are less than 10% of the whole dataset. Therefore they can be replaced with a mean value.
dataset['Total expenditure']=dataset['Total expenditure'].fillna(dataset['Total expenditure'].mean())
null_counts=null_counter(dataset.columns,dataset)
null_counts
```

```python
# 'GDP' feature is extremely correlated with 'percentage expenditure', so it should be possible to accurately predict it.
# Predicting missing 'GDP' values
dataset=dataset.reset_index(drop=True)
test_dataset=dataset[dataset['GDP'].isnull()]
train_dataset=dataset[dataset['GDP'].notnull()]
y_train=train_dataset['GDP']
X_train=train_dataset[['percentage expenditure']]
X_test=test_dataset[['percentage expenditure']]
linreg=LinearRegression()
linreg.fit(X_train,y_train)
y_pred=linreg.predict(X_test)
```

```python
# Replacing missing values
i_list=[]
for n in enumerate(dataset['GDP'].isnull()):
    if n[1]:
        i_list.append(n[0])
        m=0
for n in i_list:
    dataset['GDP'][n]=y_pred[m]
    m+=1
null_counter(dataset.columns,dataset)
```

```python
# 'Population' feature has a lot of missing values and isn't highly correlated with most of the features. Therefore it can be dropped.
dataset=dataset.drop('Population',axis=1)
null_counter(dataset.columns,dataset)
```

```python
# 'Income composition of resources' and 'Schooling' are both highly correlated with the target variable and other features. These missing values should be predicted.
# Predicting missing 'Income composition of resources' values
dataset=dataset.reset_index(drop=True)
test_dataset=dataset[dataset['Income composition of resources'].isnull()]
train_dataset=dataset[dataset['Income composition of resources'].notnull()]
y_train=train_dataset['Income composition of resources']
X_train=train_dataset.drop(['Country','Status','Total expenditure','GDP','Income composition of resources', 'Schooling'],axis=1)
X_test=test_dataset.drop(['Country','Status','Total expenditure','GDP','Income composition of resources','Schooling'],axis=1)
linreg=LinearRegression()
linreg.fit(X_train,y_train)
y_pred=linreg.predict(X_test)
```

```python
# Replacing missing values
i_list=[]
for n in enumerate(dataset['Income composition of resources'].isnull()):
    if n[1]:
        i_list.append(n[0])
        m=0
for n in i_list:
    dataset['Income composition of resources'][n]=y_pred[m]
    m+=1
null_counter(dataset.columns,dataset)
```

```python
# Predicting missing 'Schoolin' values
dataset=dataset.reset_index(drop=True)
test_dataset=dataset[dataset['Schooling'].isnull()]
train_dataset=dataset[dataset['Schooling'].notnull()]
y_train=train_dataset['Schooling']
X_train=train_dataset.drop(['Country','Status','Schooling'],axis=1)
X_test=test_dataset.drop(['Country','Status','Schooling'],axis=1)
linreg=LinearRegression()
linreg.fit(X_train,y_train)
y_pred=linreg.predict(X_test)
```

```python
# Replacing missing values
i_list=[]
for n in enumerate(dataset['Schooling'].isnull()):
    if n[1]:
        i_list.append(n[0])
        m=0
for n in i_list:
    dataset['Schooling'][n]=y_pred[m]
    m+=1
null_counter(dataset.columns,dataset)
```

```python
# Encoding categorical variables
country=pd.get_dummies(dataset['Country'],drop_first=True)
status=pd.get_dummies(dataset['Status'],drop_first=True)
dataset=dataset.drop(['Country','Status'],axis=1)
dataset=pd.concat([dataset,country,status],axis=1)
dataset.sample(5)
```

```python
# Splitting the dataset into features and dependent variable dataframes
X=dataset.drop('Life expectancy ',axis=1)
y=dataset['Life expectancy ']
```

```python
# Since the predicted value is numerical and categorical features are present, Linear Regressors are the best choice
# Feature scaling is required
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
X=std.fit_transform(X)
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
```

```python
# Performing the train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
# Linear Regression
lr=LinearRegression()
from sklearn.model_selection import cross_val_score
score_lr=cross_val_score(lr,X_train,y_train,scoring='neg_mean_squared_error')
print(score_lr.mean())
```

```python
# Lasso regression (max_iter higher than 10000 takes too much to load)
ls=Lasso(alpha=0.00033,max_iter=10000)
score_ls=cross_val_score(ls,X_train,y_train,scoring='neg_mean_squared_error')
print(score_ls.mean())
```

```python
# Ridge regression
rg=Ridge()
score_rg=cross_val_score(rg,X_train,y_train,scoring='neg_mean_squared_error')
print(score_rg.mean())
```

```python
# Elastic Net regression (max_iter higher than 10000 takes too much to load)
el=ElasticNet(alpha=0.00025,max_iter=10000)
score_el=cross_val_score(el,X_train,y_train,scoring='neg_mean_squared_error')
print(score_el.mean())
```

```python
# Predicting with the highest score regressor
el.fit(X_train,y_train)
y_pred=el.predict(X_test)
```

```python
# Measuring the accuracy of the model
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print(r2)
```
